```python
def sigmoid(hx):
    return 1.0/(1.0+np.exp(-1*hx))
```

```python
def hypothesis(w,x):
    return sigmoid(np.dot(w,x))
```

```python
hypothesis([1,1,1,1],x[0])
```

```python
def liklihood(w,x,y):
    l = 0.0
    for i in range(x.shape[0]):
        gx = hypothesis(w,x[i])
        l += y[i]*np.log2(gx) + (1-y[i])*np.log2(1-gx)
    return l/x.shape[0]
```

```python
def gradient(w,x,y):
    grad = np.zeros(w.shape[0])
    for i in range(x.shape[0]):
        gx = hypothesis(w,x[i])
        grad += (y[i] - gx)*x[i]
    return grad/x.shape[0]
```

```python
def gradient_ascent(x,y,learning_rate = 0.1):
    l_list = []
    w_list = []
    w = np.zeros(x.shape[1])
    for i in range(1000):
        l = liklihood(w,x,y)
        grad = gradient(w,x,y)
        l_list.append(l)
        w_list.append(w)
        w += learning_rate*grad
    return w,l_list,w_list
```

```python
weight,like_list,weight_list = gradient_ascent(x,y)
```

```python
plt.plot(like_list)
plt.show()
```

```python
def predict(w,x):
    confidence = hypothesis(w,x)
    if confidence < 0.5:
        return 0
    else:
        return 1
```

```python
def accuracy(w,x,y):
    y_pred = []
    for i in range(x.shape[0]):
        y_pred.append([predict(w,x[i])])
    y_pred = np.array(y_pred)
    print(y_pred,y)
    return float((y_pred==y).sum())/x.shape[0]
```

```python
accuracy(weight,x,y)
```
